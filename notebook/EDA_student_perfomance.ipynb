### Adding columns to create "Total Score" and "Average Score"

```python
df['total_score'] = df['math_score'] + df['reading_score'] + df['writing_score']
df['average_score'] = (df['total_score']/3).round(2)
```

```python
df.head()
```

```python
reading_full = df[df['reading_score'] == 100]['average_score'].count()
writing_full = df[df['writing_score']==100]['average_score'].count()
math_full = df[df['math_score']==100]['average_score'].count()

print(f'Number of students with full marks in Maths: {math_full}')
print(f'Number of students with full marks in Reading: {reading_full}')
print(f'Number of students with full marks in Writing: {writing_full}')
```

```python
reading_less_20 = df[df['reading_score']<= 20]['average_score'].count()
writing_less_20 = df[df['writing_score']<= 20]['average_score'].count()
maths_less_20 = df[df['math_score']<= 20]['average_score'].count()

print(f'Number of students with less than 20 marks in Maths: {maths_less_20}')
print(f'Number of students with less than 20 marks in Reading: {reading_less_20}')
print(f'Number of students with less than 20 marks in Writing: {writing_less_20}')

```

### Insights 

* From above values we get students have performed the worst in Maths
* Student performed best  in Reading

```python
plt.figure(figsize=(10,8))

X =['Total Average', 'Maths Average']

female_scores = [gender_groupp['average_score'][0], gender_groupp['math_score'][0]]
male_scores = [gender_groupp['average_score'][1], gender_groupp['math_score'][1]]

X_axis = np.arange(len(X))

plt.bar(X_axis - 0.2, male_scores, 0.4, label = 'Male')
plt.bar(X_axis + 0.2, female_scores, 0.4, label = 'Female')

plt.xticks(X_axis, X)
plt.ylabel('Scores')
plt.title('Total Average vs Maths Average', fontweight='bold')
plt.legend()
plt.show()
```

### 4.4.5 TEST PREPARATION COURSE COLUMN
+ Which type of lunch is most common amoung students ?
+ Is Test prepration course has any impact on student's performance ?

### BIVARIATE ANALYSIS ( Is Test prepration course has any impact on student's performance ? )

```python
# Set style for better aesthetics
sns.set_style("whitegrid")
plt.figure(figsize=(14,8))

# Define color palette
palette = sns.color_palette("pastel")

# First subplot: Math Score
plt.subplot(2,2,1)
sns.barplot(x=df['lunch'], y=df['math_score'], hue=df['test_preparation_course'], palette=palette)
plt.title("Math Score vs Lunch Type", fontsize=12)
plt.xlabel("Lunch Type")
plt.ylabel("Math Score")

# Second subplot: Reading Score
plt.subplot(2,2,2)
sns.barplot(x=df['lunch'], y=df['reading_score'], hue=df['test_preparation_course'], palette=palette)
plt.title("Reading Score vs Lunch Type", fontsize=12)
plt.xlabel("Lunch Type")
plt.ylabel("Reading Score")

# Third subplot: Writing Score
plt.subplot(2,2,3)
sns.barplot(x=df['lunch'], y=df['writing_score'], hue=df['test_preparation_course'], palette=palette)
plt.title("Writing Score vs Lunch Type", fontsize=12)
plt.xlabel("Lunch Type")
plt.ylabel("Writing Score")

# Adjust layout for better spacing
plt.tight_layout()
plt.show()
```

### Insights
+ Students who have completed the Test Prepration Course have scores higher in all three categories than those who haven't taken the course
